```python
import pandas as pd
from pathlib import Path
```

```python
# Load data from local files
historical_data_path = Path("cleaned_aqi_weather_dataset.csv")
predictions_path = Path("prediction_results.csv")

# Load historical dataset
if not historical_data_path.exists():
    raise FileNotFoundError(f"Historical data file not found: {historical_data_path}. Please ensure cleaned_aqi_weather_dataset.csv exists in the project root.")

historical_data = pd.read_csv(historical_data_path)
print(f" Loaded historical data from {historical_data_path}")
print(f"   Shape: {historical_data.shape}")

# Load predictions (if available)
if predictions_path.exists():
    predictions = pd.read_csv(predictions_path)
    print(f" Loaded predictions from {predictions_path}")
    print(f"   Shape: {predictions.shape}")
else:
    print(f" Predictions file not found: {predictions_path}")
    print("   Run prediction.ipynb first to generate predictions.")
    predictions = None
```
